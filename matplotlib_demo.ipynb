```python
# OO Style
fig=plt.figure() # instantiate a canvas

axes=fig.add_axes([0.1,0.1,0.8,0.8]) #creates a blank graph
axes.plot(x,y_cos) # first plot the data and then label axis and title
axes.set_xlabel('X label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# subplots using Functional methods
# #create a empty figure (i.e. canvas)
fig=plt.figure(figsize=(8,2))
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #first graph, list has distance from origin and height and width %
axes2=fig.add_axes([0.2,0.15,0.4,0.3]) #second graph

axes1.plot(x,y_sin)
axes1.set_title('LARGER PLOT')


axes2.plot(x,y_cos)
axes2.set_title('SMALLER PLOT')
```

```python
# subplots using Functional methods
# use array index to tell which plot has to be changed

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,2))
axes[0].plot(x,y_cos)
axes[0].set_title('First Plot')

axes[1].plot(x,y_sin)
axes[1].set_title('Second Plot')


plt.tight_layout()
```

```python
# subplots using Functional methods
# use for loop to iterate through the plots

fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y_cos)
plt.tight_layout()
plt.show()
```

```python
# OO style
## add legends
## add two lines to the same plot and differentiate by legend

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x_square')
ax.plot(x,x**3,label='x_cube')
ax.legend(loc=0) # specify legend location to avoid overlap 0 means top left
plt.show()
```
